O código de operação é composto por 8 bits, o que permite um total de 256 instruções diferentes. Dado que existem 3 tipos de instruções, cada tipo terá aproximadamente 85 instruções.

Para o tipo A (célula 0 a 84), cada instrução tem 16 bits, sendo 8 bits para o código de operação e 8 bits para o operando. Como cada célula tem 1 byte (8 bits), são necessárias 2 células para armazenar uma instrução. Portanto, o contador de instruções (CI) deve ser incrementado de 2 em 2: CI <- CI + 2.

Para o tipo B (célula 85 a 170), cada instrução tem 32 bits, o que equivale a 4 células. Nesse caso, o CI deve ser incrementado de 4 em 4: CI <- CI + 4.

Para o tipo C (célula 171 a 256), cada instrução tem 48 bits, o que corresponde a 6 células. O CI deve ser incrementado de 6 em 6: CI <- CI + 6.

Essas atualizações do CI garantem que o próximo endereço de memória a ser acessado corresponda à próxima instrução válida, considerando o tipo de instrução e seu tamanho.

R=
Add Op. - 2 ciclos de leitura da memória: Durante o primeiro ciclo, é lida a instrução do endereço indicado pela operação "Add Op.". Essa instrução contém o código de operação e o endereço do valor a ser somado com o conteúdo de R0. No segundo ciclo, ocorre a leitura do valor armazenado no endereço indicado pelo operando, que será somado ao valor de R0.

Sub Op. - 2 ciclos de leitura da memória: No primeiro ciclo, é lida a instrução do endereço indicado pela operação "Sub Op.". Essa instrução contém o código de operação e o endereço do valor a ser subtraído do conteúdo de R0. No segundo ciclo, ocorre a leitura do valor armazenado no endereço indicado pelo operando, que será subtraído do valor de R0.

Add Op1, Op2 - 4 ciclos de Memória: No primeiro ciclo, ocorre a leitura da instrução que contém os códigos de operação e os endereços dos operandos. No segundo ciclo, é feita a leitura do conteúdo do primeiro operando. No terceiro ciclo, é feita a leitura do conteúdo do segundo operando. Esses dois operandos são somados dentro da Unidade Lógica e Aritmética (UAL). No quarto ciclo, o resultado dessa soma é escrito na célula de memória equivalente ao primeiro operando.

INCR - 1 Ciclo de memória: A instrução "INCR" é lida em um ciclo de memória, e imediatamente o valor de R0 é incrementado.

LDA OP - 2 ciclos de leitura da memória: No primeiro ciclo, é feita a leitura da instrução que contém o código de operação e o endereço do operando a ser carregado em R0. No segundo ciclo, é feita a leitura do conteúdo do endereço indicado pelo operando, que será carregado em R0.

R=
CI=16: O tamanho de bits do CI pode ser determinado considerando que o CI é responsável por indicar o próximo endereço de instrução. Portanto, faz sentido que ele tenha o mesmo tamanho do endereço de memória. No caso, o tamanho dos endereços é determinado pela quantidade de células de memória disponíveis, que é 64K ou 2^16. Isso significa que o tamanho dos endereços e do CI é igual a 16 bits. Assim, o campo de operando também terá 16 bits, já que é o mesmo tamanho do CI.

Reg(ACC)= 24: O tamanho de bits do Reg pode ser encontrado considerando que ele armazena as instruções. Nesse caso, o tamanho do Reg será determinado pela soma do campo de operando com o tamanho do código de operação.

Já determinamos que o campo de operando possui 16 bits. Agora, vamos considerar a quantidade de instruções disponíveis, que é 256 ou 2^8. Isso significa que o código de operação terá 8 bits. Somando esses 8 bits com os 16 bits do campo de operando, obtemos um total de 24 bits para o Reg (ACC).

RDM (Registrador de Dados de Memória): O RDM também armazena instruções, portanto, terá o mesmo tamanho que o Reg (ACC), ou seja, 24 bits.

O total de bits na memória é determinado multiplicando o tamanho da instrução (célula e palavra) pela quantidade de células disponíveis. Sabemos que cada instrução tem 24 bits, e a memória possui 64K células. Portanto, o total de bits na memória será 24 bits multiplicado por 64K instruções, resultando em 1,5 milhão de bits.

```python
import math
# calculo do CI
numero_celulas = 64 * 1024  # 64K células de memória

bits_operando = math.log2(numero_celulas)

print(f"Tamanho em bits do CI : {bits_operando}")

# calculo do ACC

campo_operando = 16
cod_operacao = 8

tamanho_reg = campo_operando + cod_operacao
print("Tamanho do Reg (ACC):  ", tamanho_reg, "bits")

print("Tamanho do RDM :       ", tamanho_reg, "bits também")

# total de bits
tamanho_instrucao = 24
quantidade_celulas = 64 * 1024  # 64K

total_bits = tamanho_instrucao * quantidade_celulas
print("Total de bits na memória:", total_bits)
print("OU 1,5 M bits")
```
